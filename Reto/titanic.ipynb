#### **CONSULTANDO Y ELIMINANDO DUPLICADOS**

```python
duplicados = datos.duplicated().sum()
if duplicados > 0:
    print(f"Se encontraron {duplicados} filas duplicadas. Procediendo a eliminarlas")
    datos.drop_duplicates(inplace=True)
else:
    print("No se encontraron filas duplicadas. No es necesario eliminarlas")

print("Dimensiones tras el paso de duplicados:", datos.shape)
```

#### **IDENTIFICANDO Y ELIMINANDO NULOS**
isnull = Identifica los nulos ////
dropna = Elimina los nulos

```python
nulos_columnas = datos.isnull().sum()
nulos_filas = datos[datos.isnull().any(axis=1)]
total_nulos = nulos_columnas.sum()
```

```python
nulos_columnas
```

```python
nulos_filas
```

```python
print(f" El total de nulos es: {total_nulos}")
```

**IMPUTACION Y ELIMINACION**

```python
if nulos_columnas['Age'] > 0:
    datos['Age'] = datos['Age'].fillna(datos['Age'].median())
    print("Reemplazo de nulos en la mediania!")
else:
    print("No hay valores nulos que limpiar")

if nulos_columnas['Embarked'] > 0:
    datos.dropna(subset=['Embarked'], inplace=True)
    print("Limpieza de nulos completada!")
else:
    print("No hay valores nulos que limpiar")

if nulos_columnas['Survived'] > 0:
    datos.dropna(subset=['Survived'], inplace=True)
    print("Limpieza de nulos completada!")
else:
    print("No hay valores nulos que limpiar")
```

#### **GRAFICO DE SUPERVIVIENTES VS FALLECIDOS**

#### GRAFICO DE BARRA (TERMINADO(creo))

```python
conteo = datos['Survived'].value_counts()
etiquetas = ['Fallecidos', 'Supervivientes']
colores = ['red', 'green']


plt.bar(etiquetas, conteo, color=colores)
plt.title('Supervivientes vs Fallecidos')
plt.xlabel('Estado')
plt.ylabel('Cantidad de personas')

plt.show()

```

#### GRAFICO CIRCULAR (NO TERMINADO)

```python
promedioSUP = datos[['Survived']]
promSUP = promedioSUP.mean()
```

```python
promedioFALL = datos[['Survived']]
promFALL = promedioFALL.mean()
```

```python
dataset_promedios = pd.DataFrame({'Survived': ['1', '0'], 'Promedio':[promSUP, promFALL]})
```

```python
plt.pie(dataset_promedios['Promedio'], labels=dataset_promedios['Survived'], autopct='%1.1f%%')
plt.title("SUPERVIVIENTES VS FALLECIDOS")
```
