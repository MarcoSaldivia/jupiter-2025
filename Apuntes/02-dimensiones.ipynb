```python
import numpy as np
```

#### **Generando un Vector Aleatorio de 12 numeros enteros (desde el 1 al 10 ) con Random**

```python
vector = np.random.randint(1, 11, 12) # Tercer parametro se coloca la cantidad de elementos que tendra el vector
vector
```

#### **Obtencion de la Dimension de un Arreglo**

```python
np.shape(vector)
```

#### **Transformar el vector de 12 elementos en una Matriz 4x3**

```python
#El metodo reshape() redimensiona una Estructura
matriz = vector.reshape(4,3)
matriz
```

#### **Otra variante de matriz**

```python
matriz2 = vector.reshape(6,2)
matriz2
```

#### **Volviendo a consultar la dimension pero de la Matriz**

```python
np.shape(matriz)
```

### **Conociendo la Biblioteca Matplotlib**

El paquete **matplotlib** permite crear visualizaciones (graficos) para interpretar resultados y data

```python
import matplotlib.pyplot as plt
```

#### **Creado un Grafico Simple**

```python
plt.plot(vector, label = "Vector")
plt.title("Grafico del Vector")
plt.xlabel("Indice (Posicion)") # Agrega el titulo de eje x
plt.ylabel("Valores") # Agrega el titulo de eje y

plt.grid(True) # Para agregar una grilla al grafico
plt.legend() # Etiqueta, que se obtiene del label del metodo plot()

plt.show() # Permite visualizar el grafico
```

#### **Crear 2 Vectores Aleatorios en el mismo Grafico**

```python
vector1 = np.random.randint(1, 11, 12)
vector2 = np.random.randint(2, 21, 12)
```

```python
plt.plot(vector1, marker = "o", label = "Vector 1") # La popiedad marker, agrega un marcador en los valores del vector
plt.plot(vector2, marker = "o", label = "Vector 2")

plt.title("Comparando 2 Vectores")
plt.xlabel("Indice (Posicion)")
plt.ylabel("Valores")

plt.grid(True)
plt.legend()

plt.show()
```

#### **Creando un Grafico de Barra**

Un grafico de barra con los Vectores antes mencionados

```python
plt.bar(range(len(vector2)), vector1) # El metodo bar, permite plotear un grafico de barra
```
