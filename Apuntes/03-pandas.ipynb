#### **Funcion para verificar si un genero de videojuego esta presente en la lista**

En el ejemplo vamos a verificar si se encuentra el genero RPG en la lista.

```python
def filtro_rpg(genres):
    return 'RPG' in genres

# Aplicar un filtro para consultar y llamar la solucion filtro_rpg()
datosRPG = datos[datos['Genres'].apply(filtro_rpg)] # Creando un nuevo dataframe con los videojuegos RPG

# Mostrar el dataset filtrado, inclyendo solo las columnas (Titulo y Rating)
nombresRPG = datosRPG[['Title', 'Rating']]
nombresRPG.head(20) # Que se muestren los primeros 20 juegos RPG
```

#### **Consultando informacion del nuevo DataSet**

```python
nombresRPG.info()
```

#### **Graficar los primeros 10 Videojuegos RPG y su Rating de popularidad (Grafico de Barra)**

```python
graficoRPG = nombresRPG.head(10) # grafico solo tiene los primeros 10 Videojuegos

graficoRPG.drop(['Title'], axis=1)['Rating'].plot(kind='bar')
```

#### **Agregar el titulo correspondiente**

```python
graficoRPG = nombresRPG.head(10) # grafico solo tiene los primeros 10 Videojuegos

graficoRPG.set_index('Title', inplace=True) # Se establece la columna 'Title' como indice
graficoRPG['Rating'].plot(kind='bar')
```

#### **Agregando las etiquetas de datos al grafico**

```python
graficoRPG = nombresRPG.head(10) # grafico solo tiene los primeros 10 Videojuegos

graficoRPG.set_index('Title', inplace=True) # Se establece la columna 'Title' como indice
graph = graficoRPG['Rating'].plot(kind='bar')

# Agregar las etiquetas numericas de rating en barras
for i, rating in enumerate(graficoRPG['Rating']):
    graph.annotate(str(rating), xy=(i, rating), ha='center', va='bottom') # annotate agrega str o escribe en el plot

plt.show()
```

### **Grafico de Torta**

#### **Promedio de Rating de Videojuegos RPG**

```python
promedioRPG = datosRPG[['Rating']] # Promedio de Videojuegos RPG
promRPG = promedioRPG.mean()
```

#### **Promedio de Rating de Videojuegos RPG Indie**

```python
# Funcion que filtra por genero indie
def filtro_indie(genres):
    return 'Indie' in genres

# Aplicar un filtro para consultar y llamar la funcion filtro_indie()
datosIndie = datos[datos['Genres'].apply(filtro_indie)] # Creando un nuevo dataframe

# Mostrar el dataset filtrado, incluyendo solo las columnas (Titulo y Rating)
nombresIndie = datosIndie[['Title', 'Rating']]
nombresIndie.head(20) # Que se muestren los primeros 20 videojuegos indie
```

```python
promedioIndie = datosIndie[['Rating']]  # Promedio de Videojuegos RPG
promInd = promedioIndie.mean()
```

#### **Crear Grafico Circular**

```python
dataset_promedios = pd.DataFrame({'Genero': ['RPG', 'Indie'], 'Promedio':[promRPG, promInd]})
```

```python
plt.pie(dataset_promedios['Promedio'], labels=dataset_promedios['Genero'], autopct='%1.1f%%')
plt.title("Comparacion de Rating entre RPG e Indie")
```
